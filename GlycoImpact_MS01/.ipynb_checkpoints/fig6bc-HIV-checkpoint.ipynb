## Lysine ~ Highmannose check
Here we examine proximate lysines (within 5 Angstroms) for their assocaition with high mannose glycosylation sites. In this examination, we use the high-mannose/complex ratio to indicate the relative preference for high-mannose vs complex glycans; ratio>1 indicates a preference for high-mannose, ratio<1 indicates a preference for complex, ratio=1 indicates no preference.

Comparing the high-mannose/complex ratio directly, it appears that glycosylation sites with no proximal Lysine have effectively no preference for high-mannose or complex as indicated by the median of .794, nearly 1. When a Lysine is nearby, within 5A, the ratio increases dramatically to 3.2 (anova, p=0.015)

Examining the high-mannose/complex ratio in the context of hydrophobicity [Kyte & Doolittle], we see that both factors appear to influence the ratio. In this context, lysine proximity has a 1.85 increase in high-mannose preference (wald, p=0.02897) while hydrophobicity has a 1.05 increase / hydrophobic unit (wald, p=0.018). There is no decernable interaction though both variables appear to incluence mannose preference.

## Univariate Predictors of HighMannose in HIV

Fitting univariate gaussian generalized regression models, we found several more protein structural features that appear to be associated with high-mannose/complex glycan preference. The figure below shows the coefficient for each protein structure feature influence on the glycan preference. Points are colored by the -log10(p) with the p-value of the wald test indicating the probability a coefficient is non-zero. The error bars indicate 95% confidence intervals for the coefficients.

In the univariate models we find that residue depth (RES_DEPTH), surface area (RSA), and hydrophobicity are all significantly associated with glycan preference.  Only the strand (E) secondary structure is signficantly associated with high-mannose preference. These findings are consistent with our understanding of high mannosylation; an immature glycosylation event. If these residues are burrued or inaccessible, the mannoses may not be removed and the complex glycans will not be attached. 

Looking at proximate amino acids, V, T, W, G, A and nearly K are significantly associated with high-mannose peference. This is an interesting combination of hydrophobic (V,A,W), small (V,A,G) and one polar (T) and one positive (K) residue. I am curious if this array of proximal amino acids suggests two modes of high-mannosylation events: (1) high-burrial event which occurs where maturation of the glycan is prevented by interfering amino acids, (2) a surface-bonding event where maturation of the glycan is prevented by its interaction with polar or charged amino acids on the surface. 

DSSP terminology:
Code	Structure
H	Alpha helix (4-12)
B	Isolated beta-bridge residue
E	Strand
G	3-10 helix
I	Pi helix
T	Turn
S	Bend

```r
#univariate
ml=lapply(str_vars,function(x) {
    mod=NULL
    try(mod<-glm( as.formula(paste('HighMannose_N203 ~',x)) , data=m ))
#    try(mod<-glm( as.formula(paste('HighMann_Complex ~',x)) , data=m ))
    if(is.null(mod)){return(NULL)}
    conf = confint(mod,verbose=F)
    if(nrow(conf)!=nrow(coef(summary(mod)))){return(NA)}
    cbind(coef(summary(mod)),conf)
})
names(ml)=str_vars
ml = na.omit(as.data.frame(do.call(rbind,ml)))
colnames(ml) = c('coef','std.err','t','p','min','max')
#ml$q=p.adjust( ml$p,'fdr')
ml$var=rownames(ml)
```

```r
#ml$var = relevel(factor(ml$var),ml$var[order(ml$p)])
ml$var = factor(as.character(ml$var),levels=as.character(ml$var)[order(ml$coef)])
ggplot(ml[!grepl('Intercept',ml$var)&abs((ml$coef))>.1,],aes(x=var,y=coef,color=-log(p,10)))+
    geom_point()+geom_pointrange(aes(ymax=max,ymin=min))+geom_hline(yintercept = 0)+
    coord_flip()+theme_classic()+ggtitle("Magnitude and Significance of Protein Structure\nEffect on Mannose/Complex Ratio")
```

## Multidimensional Analysis

Note, 20-30 datapoints is not enough for a thorough multidimensional analysis, the following commentaries should be taken with a grain of salt.

### Principal Componant Analysis (PCA )
Principal component analysis highlight major trajectories of influence protein structure features may have on glycosylation. Here high-mannose preferring sites (blue), intermediate sites (green), and complex preferring sites (red) are mapped onto a PCA showing co-occurance of protein features. We see that Valine co-occurs with hydrophobicity and both occur orthoganally from Alanine. Therefore, while Valine appears to contribute to high-mannose selecting hydrophobicity, Alanine may be part of a different process of glycosylation. All of these trajectories, are co-occuring with depth suggesting that residue depth is a major feature of high-mannose preference. The hydrophobicity-Valine trajectory is co-occuring with Threonine proximity suggesting that burrial and direct glycan interaction may not be two seperate methods of high-mannose presevation.

### Partial Least Square Regression (PLS)
PLS shows a clear distinction between protein structure and glycan preference. While no glycan sigtes appear neglibly associated with protein structure features, complex glycans are strongly positively associated with high exposure sites near extra asparagines. High-mannose sites show the opposite effect, low exposure, no additional asparagines. In addition, high mannose sites are positively associated with several secondary structures (a helix [H], strand [E]), small or hydrophobic aa (A, G, F, V) and positive or polar aa (K, H, T) as well as high depth. All of these associations are inverted for high-complexity sites revealing a strong association between protein structure features and glycosylation

## Other charged amino acids do not appear to contribute to the Mannose/Complex Ratio

In the case of the HIV envelope, Lysine appears to be the only changed amino acid significantly and substantially associated with high-mannosylation. There is one site with a proximal histadine and extremely high mannosylation. This effect is technically significant (p=.04) but, as it is only one site, we are limited in the conclusions we can draw. 

We attempted aggregating over all positive, all negative, and all chaged amino acids (total number and binary >0). The sum of positive amino acids was significantly associated with increased high-mannose though this effect appears to be mostly due to the lysine effect. The sums of posative, negative and all charged amino acids was not significantly associated with the mannose/complex ratio. The binary presence of positive amino acids was marginally significant (p=0.046), but again, this is likely due to the strength of the lysine effect. Finally, there is no significant additive effect or interaction between the number of positive and negative proximal charged amino acids.
